# 3. Data Cleaning

3.1 Identifying the Null values

## Verifying the % of null values

```python
(loan_df.isnull().sum()/len(loan_df)*100).sort_values(ascending=False)
```

3.2 Dealing with Data

3.2.1 Removing Columns with null values

```python
loan_df2 = loan_df.loc[:, loan_df.isnull().sum()/len(loan_df)*100 <60]
```

```python
loan_df2.shape
```

re-verify null values in dataframe

```python
loan_df2.nunique()
```

3.2.2 Dealing with NAN values

```python
loan_df2.fillna(value=0,inplace=True)
```

```python
for col in loan_df2.columns:
    if len(loan_df2[col].unique())==1 :
        loan_df2.drop(col,inplace=True,axis=1)
```

```python
# id,member_id and url are unique values, but it holds id info of customers, and this variable cannot be manipulated
# so we can use one as primary key and drop rest
loan_df2.drop(['id','url'],axis=1,inplace=True)
```

```python
loan_df2.info()
```

Columns ['collection_recovery_fee','total_rec_late_fee','recoveries','total_rec_int'] are mainly focusing of loans that are in Current state
so those won't help us

```python
loan_df2.drop(['collection_recovery_fee','total_rec_late_fee','recoveries','total_rec_int'],axis=1,inplace=True)
```

```python
loan_df2.info()
```

Verify Null/NAN values in dataframe

```python
(loan_df2.isnull().sum()/len(loan_df2)*100).sort_values(ascending=False)
```

We need to mainly focus on values 'Charged-Off' and 'Fully Paid' as 'Current' status means Loan is still in progress

```python
# removing the rows with loan status 'Current'
loan_df_non_current_ls = loan_df2[loan_df2['loan_status']!='Current']
```

```python
# make sure we have only values that are required in loan status
loan_df_non_current_ls['loan_status'].unique()
```

```python
loan_df_non_current_ls['earliest_cr_line']= loan_df2['earliest_cr_line']
```

columns 'int_rate','revol_util' are defined in Percentage. It's better to convert them to numeric values

```python
loan_df_non_current_ls['int_rate']= loan_df_non_current_ls['int_rate'].str.strip('%').astype(float)
```

```python
loan_df_non_current_ls['revol_util']= loan_df_non_current_ls['revol_util'].str.strip('%').astype(float)
```

```python
# Convert issue_d from MM-YY to YY-mm-dd and make it datetime column
loan_df_non_current_ls['issue_d'] = pd.to_datetime(loan_df_non_current_ls['issue_d'],format='%b-%y')
```

```python
# Convert earliest_cr_line from MM-YY to DD-mm-yyyy
loan_df_non_current_ls['earliest_cr_line'] = pd.to_datetime(loan_df_non_current_ls['earliest_cr_line'],format='%b-%y')
```
